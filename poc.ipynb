```python
starts = concatenated_df.groupby('start_station_id')
ends = concatenated_df.groupby('end_station_id')
```

```python
sdf = starts['start_station_name'].agg(['count'])
```

```python
sdf.columns = ['start_rides']
```

```python
sdf.head()
```

```python

sdf['end_rides'] = ends['start_station_name'].agg(['count'])
```

```python
sdf[['lat', 'lon','station_name' ]] = starts[
    ['start_station_latitude', 'start_station_longitude', 'start_station_name']].agg(['min'])
```

```python
 m = folium.Map(location=[40.765, -73.9], zoom_start=12)
```

```python
from folium.plugins import MarkerCluster
marker_cluster = folium.plugins.MarkerCluster().add_to(m)
```

```python
for i, row in sdf.iterrows():
    folium.Marker(
        location=[row.lat, row.lon],
        popup = row.station_name).add_to(marker_cluster) 
```
